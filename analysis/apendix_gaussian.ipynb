# Apendix: Gauss Smoothing example

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, mu, sigma):
    return (1.0 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-((x - mu)**2) / (2 * sigma**2))
```

```python
fig = plt.figure()
ax = plt.axes()

# Plot vertical lines
for x in np.arange(0, 1, (1.0 / 30.0)):
    if x == 0.5:
        color = 'red'
    else:
        color = '#0f0f0f50'
    plt.axvline(x=x, color=color)

# Plot gaussian distribution
x = np.arange(0, 1, 1.0 / 300.0)
for i in range(1, 5):
    y = gaussian(x, 0.5, i / 30.0)
    y = y / np.sum(y)
    ax.plot(x, y, label=f'$\sigma={i} \\times P$')
    
fig.tight_layout()
plt.legend()
plt.show()
```

```python
P = 1.0 / (1.0 / 30.0)
x = np.arange(5) * P
y = gaussian(x, x[-1], P)
y = y / np.sum(y)
print([f'{i:.5f}' for i in y])
```
